# Limits and Continuity

Limits are essential to calculus and mathematical analysis, and are used to
define continuity, derivatives, and integrals.

Let $f : D \to \mathbb{R}$ be a real-valued function on $D \subseteq \mathbb{R}^d$. Then $f$ is said to have a limit $L \in\mathbb{R}$ as $\textbf{x}$ approaches $\textbf{a}$ if: for any $\epsilon > 0$, there exists a $\delta > 0$ such that $|f(\textbf{x})−L| < \epsilon$ for all $\textbf{x} \in D\cap B_\delta(\textbf{a})\backslash\{\textbf{a}\}$. This is written as
$$\lim_{\textbf{x}\to \textbf{a}}f(\textbf{x}) = L.$$

**Example 1**<br>
$\lim_{x\to 0}\frac{\sin(x)}{x}$

```python
import sympy as sp
x=sp.Symbol('x')
f=sp.sin(x)/x
L=sp.limit(f,x,0)
print("The Limit is:",L)

```

**Example 2**<br>
$\lim_{x\to 0}\frac{\sin(2y)}{y}$

```python
y=sp.Symbol('y')
f=sp.sin(2*y)/y
L=sp.limit(f,y,0)
print("The limit is:",L)
```

# Composition of Continuous Functions

The composition of a function is an operation where two functions say f and g generate a new function say h in such a way that h(x) = g(f(x)) as shown below.



```python
# Function to combine two
def composite_function(f, g):
    return lambda x : f(g(x))

# Function to add 6
def add(x):
    return x + 6

# Function to multiply 2
def multiply(x):
    return x * 2

# Composite function returns a lambda function
add_multiply = composite_function(multiply, add)

print("Adding 6 to 5 and multiplying the result with 2: ",
      add_multiply(5))
```

# Derivatives

The derivative of a function of a real variable measures the sensitivity to
change of the function value (output value) with respect to another variable.

 Let $f : D \to\mathbb{ R}$ where $D \subseteq \mathbb{R}$ and let $x_0 \in D$ be an interior point of $D$. The derivative of $f$ at $x_0$ is
 $$f′(x_0) = \frac{d f(x_0)}{dx}= \lim_{h\to 0}\frac{f(x_0 +h)− f(x_0)}{h}$$
provided the limit exists.

**Example 3**<br>
$f(x)=\sin(x)+x^2+e^{4x}$

```python
import sympy as sp
x=sp.Symbol('x')
f=sp.sin(x)+x**2+sp.exp(4*x)
print(f)
df=sp.diff(f,x)
print(df)
ddf=sp.diff(f,x,2)
print(ddf)
```

**Example 4**<br>
$f(y)=\cos(y)+4y^2+y^{3}$

```python
y=sp.Symbol('y')
f=sp.cos(y)+4*y+y**3
print(f)
df=sp.diff(f,y)
print(df)
```

# Mean-Value Theorem

The mean value theorem states, roughly, that for a given planar arc between two endpoints, there is at least one point at which the tangent to the arc is parallel to the secant through its endpoints. It is one of the most important results in real analysis.

Let $f : [a,b] \to\mathbb{ R}$ be a continuous function and
assume that its derivative exists on $(a,b)$. Then there is $a < c < b$ such that
$$f(b) = f(a) + (b−a)f'(c),$$
or put differently
$$\frac{f(b)− f(a)}{b−a}= f'(c).$$

#3.2.3 Taylor’s Theorem

Taylor’s theorem gives an approximation of a differentiable function around
a given point by a polynomial. It is a powerful generalization of the Mean Value Theorem that provides polynomial approximations to a function around a point.


Let's take the example of the function $f(x) = e^{2}$. The actual value of f(x) is

```python
print(math.exp(2))
```

Calculating the sum of the first five terms of the Taylor Series expansion gives us an approximate calculation for f(x).

```python

import math

x = 2
e_to_2 = x**0/math.factorial(0) + x**1/math.factorial(1) + x**2/math.factorial(2) + x**3/math.factorial(3) + x**4/math.factorial(4)
print(e_to_2)
```

Increasing the number of terms increases the accuracy of our approximation.

```python
def func_e(x, n):
    e_approx = 0
    for i in range(n):
        e_approx += x**i/math.factorial(i)

    return e_approx

x = 5
for i in range(1,11):
    e_approx = func_e(x,i)
    e_app = "{:.5f}".format(e_approx)
    e_exp = math.exp(x)
    e_ex = "{:.5f}".format(e_exp)
    e_error = abs(e_approx - e_exp)
    print(f'{i} terms: Taylor Series approx= {e_app},\t\t exp calc= {e_ex},\t\t error = {e_error}')
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline

def func_cos(x, n):
    cos_approx = 0
    for i in range(n):
        coef = (-1)**i
        num = x**(2*i)
        denom = math.factorial(2*i)
        cos_approx += ( coef ) * ( (num)/(denom) )

    return cos_approx

angles = np.arange(-6*np.pi,6*np.pi,0.1)
p_cos = np.cos(angles)

fig, ax = plt.subplots()
ax.plot(angles,p_cos)

# add lines for between 1 and 6 terms in the Taylor Series
for i in range(1,10):
    t_cos = [func_cos(angle,i) for angle in angles]
    ax.plot(angles,t_cos)

ax.set_ylim([-4,4])
ax.set_xlim([-20,40])

# set up legend
legend_lst = ['cos() function']
for i in range(1,10):
    legend_lst.append(f'Taylor Series - {i} terms')
ax.legend(legend_lst, loc=1)

plt.show()
```
